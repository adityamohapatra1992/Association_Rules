```python
from apyori import apriori
```

```python
X = apriori(product['Transaction'],min_support=0.0045,min_confidence=0.2,min_lift=2,max_len=2)
X
```

```python
Y = list(X)
Y
```

```python
for item in Y:
    pair = item[0]
    items = [ x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print('==========================')
```

```python
A = product['Transaction'].str.join(sep='*').str.get_dummies(sep='*')
```

```python
frequent_itemset = apriori(A,min_support=0.005,use_colnames=True,max_len=None)
frequent_itemset
```

```python
rules = association_rules(frequent_itemset,metric='lift',min_threshold=1)
rules.head()
```

```python
# Eliminating Redudancy

def to_list(i):
    return (sorted(list(i)))
```

```python
ma_X = rules.antecedents.apply(to_list)+rules.consequents.apply(to_list)
ma_X
```

```python
ma_X = ma_X.apply(sorted)
ma_X 
```

```python
rules_sets = list(ma_X)
rules_sets
```

```python
unique_rules_sets = [list(m) for m in set(tuple(i) for i in rules_sets)]
index_rules = []
for i in unique_rules_sets:
    index_rules.append(rules_sets.index(i))
```

```python
rules_no_redudancy  = rules.iloc[index_rules,:]
rules_no_redudancy
```

```python
rules_no_redudancy.sort_values('lift',ascending=False)
```
